```python
target = drugdset['pIC50']
features = drugdset.drop(['SMILES', 'pIC50','MolecularFormula','InChI','InChIKey','IUPACName'],axis=1)
target.shape, features.shape
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()

features_scaled = min_max_scaler.fit_transform(features)
features_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(features_scaled, target, test_size=0.3, random_state=22,)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import xgboost as xgb
import numpy as np

# Data Preparation
dtrain = xgb.DMatrix(X_train,y_train)

# Training a Model
params = {
    'objective': 'reg:squarederror',
    'max_depth': 3,
    'eta': 0.1,
    'gamma': 0.0,
    'subsample': 1.0,
    'colsample_bytree': 1.0,
    'eval_metric': 'rmse'
}
num_rounds = 10
model = xgb.train(params, dtrain, num_rounds)

# Making Predictions
dtest = xgb.DMatrix(X_test)
predictions = model.predict(dtest)

print(predictions)
```

```python
from sklearn import tree

# Data Preparation


# Creating and Training a Decision Tree Model
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)

# Making Predictions
predictions = clf.predict(X_test)

print(predictions)

# Visualizing the Decision Tree (Optional)
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("decision_tree")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline(steps=[('impute', SimpleImputer(strategy='mean'))])
def prepare_model(algorithm, X_train, y_train): 
    model = Pipeline(steps=[('preprocessing', num_pipeline),('algorithm', algorithm)])
    model.fit(X_train, y_train)
    return model
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor, BaggingRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor, ExtraTreeRegressor
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.neighbors import KNeighborsRegressor
import time
from sklearn.metrics import mean_squared_error, mean_absolute_error
algorithms = [RandomForestRegressor(), AdaBoostRegressor(), GradientBoostingRegressor(), 
              BaggingRegressor(), SVR(), DecisionTreeRegressor(), ExtraTreeRegressor(), 
              LinearRegression(), SGDRegressor(), KNeighborsRegressor()]
names = []
times = []
mses = []
maes = []

for algorithm in algorithms:
    name = type(algorithm).__name__
    names.append(name)
    print(name)
    start_time = time.time()
    model = prepare_model(algorithm, X_train, y_train)
    pred = model.predict(X_test)
    end_time = time.time()
    print(end_time - start_time)
    times.append(end_time - start_time)
    mses.append(mean_squared_error(y_test, pred))
    maes.append(mean_absolute_error(y_test, pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=2212)

dtr.fit(X_train, y_train)
print(dtr.score(X_train, y_train))
print(dtr.score(X_test, y_test))

dtr_pred = dtr.predict(X_test)
dtr_mse = mean_squared_error(y_test, dtr_pred)
print(dtr_mse)
```

```python
results_dict = {'Algorithm': names, 'MSE': mses, 'MAE': maes, 'Time': times}
pd.DataFrame(results_dict).sort_values(by='MSE', ascending=1)
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

# Example dataset
smiles_list = drugdset['SMILES']

# Calculate molecular descriptors for each compound
descriptors = []
for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        descriptor = {}
        descriptor['SMILES'] = smiles
        descriptor['MolecularWeight'] = Descriptors.MolWt(mol)
        descriptor['XLogP'] = Descriptors.MolLogP(mol)
        descriptor['TPSA'] = Descriptors.TPSA(mol)
        # Add more descriptors as needed
        descriptors.append(descriptor)

# Convert the descriptors to a pandas DataFrame
import pandas as pd
df_descriptors = pd.DataFrame(descriptors)

# Print the calculated descriptors
print(df_descriptors)
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

# Construct molecules from SMILES
molecules = []
for smiles in drugdset['SMILES']:
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        molecules.append(mol)

# Generate and display 3D diagrams of the molecules
for mol in molecules:
    Chem.Compute2DCoords(mol)  # Compute 2D coordinates
    drawer = Draw.MolDraw2DC(400, 400)  # Create a 2D drawing object
    drawer.DrawMolecule(mol)  # Draw the molecule in 2D
    drawer.FinishDrawing()  # Finish drawing
    image = drawer.GetDrawingText()  # Get the drawing as an image
    display(image)  # Display the image
```
